This is an exploratory analysis on a Uk accident data set

# Step 1 : import of necessary library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Step 2 :loading and reading of the dataset

```python
data = pd.read_csv("UK_Accident.csv")
```

# Step 3 : checking the describtion of the data

```python
data.info()
```

```python
data.tail()
```

# step 7 : Dropping the null values

```python
data.dropna(inplace = True)
```

# step 8 : loading the visualization libraries and plots

# histogram plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize = (8,4))
sns.histplot(data['Number_of_Casualties'], kde=True, bins=20, color = 'green')
plt.xlabel("Totalcasualties")
plt.ylabel("Frequency")
plt.title("Number of cassualties Distribution")
plt.show()
```

step 9 : Box Plot

```python
plt.figure(figsize = (8, 8))
sns.boxplot(x = "Number_of_Casualties", y ="Year", data=data, color="blue")
plt.xlabel("Number of casualties")
plt.ylabel("Year")
plt.title("Casualties distrubution Yearly")
plt.show()
```
