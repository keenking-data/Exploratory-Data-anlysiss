```python
#Convert posting_date and application_deadline to datetime
data['posting_date'] = pd.to_datetime(data['posting_date'])
data['application_deadline'] = pd.to_datetime(data['application_deadline'])
```

```python
#summRY STATISTICS
print("\nSummary Statistics: ")
data.describe()
```

```python
#Top 15 hot jobs by frequency
top_jobs = data['job_title'].value_counts().head(15)
print("\nTop 15 Job Tittles by frequency")
print(top_jobs)
```

```python
#Average Salary per experience level
avg_salary_by_exp = data.groupby('experience_level')['salary_usd'].mean().sort_values()
print("\nAverage Salary by Experience Level (USD) :")
print(avg_salary_by_exp)
```

```python
#No of remote jobs listed
remote_per = data['remote_ratio'].value_counts(normalize=True) * 100
print("\nRemote Work Distribution (%) :")
print(remote_per)
```

```python
# 3. Visualizations
# Plot 1: Bar plot of top 10 job titles
plt.figure(figsize=(10, 8), facecolor='black')
ax = top_jobs.plot(kind = 'bar', color = 'skyblue')
ax.set_facecolor('black')

# Set text and label colors
plt.title("Top AI Job Titles", color ='white')
plt.xlabel("Job Tittle", color = 'white')
plt.ylabel("Number of Jobs", color ='white')

# Set tick label colors
plt.xticks(rotation=45,color='white', ha='right')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['salary_usd'], bins=20, kde=True, color='black')
plt.title('Salary Distribution for AI Jobs (USD)')
plt.xlabel('Salary (USD)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Histogram of salaries
plt.figure(figsize=(10, 6), facecolor = 'black')
sns.histplot(data['salary_usd'], bins= 20, kde=True, color='white').set_facecolor('black')
plt.title('Salary Distribution for AI Jobs (USD)', color='white')
plt.xlabel('Salary (USD)', color='white')
plt.ylabel('Frequency', color='white')
plt.tight_layout()
plt.show()
```

```python
#Box plot for salary based on Experience level
plt.figure(figsize = (12, 6))
sns.boxplot(x='experience_level', y='salary_usd', data=data, order=['EN','MI','SE','EX'])
plt.title('Salary by Experience Level')
plt.xlabel('Experience Level (EN=Entry, MI=Mid, SE=Senior, EX=Executive)')
plt.ylabel('Salary (USD)')
plt.tight_layout()
plt.show()
```

```python
#Count plot of jobs by industry
plt.figure(figsize=(12, 6))
sns.countplot(y='industry', data=data, order=data['industry'].value_counts().index[:10])
plt.title('Top 10 Industries for AI Jobs')
plt.xlabel('Number of Jobs')
plt.ylabel('Industry')
plt.tight_layout()
plt.show()
```

```python
# Interesting Insight: Correlation between years of experience and salary
plt.figure(figsize=(10, 6))
sns.scatterplot(x='years_experience', y='salary_usd', hue='experience_level', data=data)
plt.title('Salary vs Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary (USD)')
plt.tight_layout()
plt.show()
print("\nInsight: Executive-level roles (EX) often have higher salaries, but some senior-level (SE) roles with 5-10 years of experience can match or exceed entry-level executive salaries, suggesting skill specialization or industry demand impacts pay significantly.")
```
