```python
import pandas as pd
df=pd.read_csv('bra.csv')
df.tail(14)
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
df['rating'].fillna(df['rating'].mean(), inplace=True)
df['initial_price'].fillna(df['initial_price'].mean(), inplace=True)
df.info()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['final_price'].value_counts().head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
# Rating distribution
sns.histplot(df['rating'], bins=10, kde=True, color='orange')
plt.title('Rating Distribution')
plt.show()
```

```python
#scatter plot final price vs rating 
sns.scatterplot(x='final_price', y='rating', data=df)
plt.title('Final Price vs Rating')
plt.show()
```

```python
# Bar chart: average rating per category
plt.figure(figsize=(10, 6))
sns.barplot(x='final_price', y='rating', data=df, color='skyblue')
plt.title('Average Rating per Final Price')
plt.xlabel('Final Price')
plt.ylabel('Average Rating')
plt.show()
```

```python
df['final_price'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['#66b3ff','#99ff99','#ffcc99'])
plt.title('Final Price Market Share')
plt.ylabel('')
plt.show()
```

```python
sns.lineplot(x='final_price', y='rating', data=df, color='orange')
plt.title('Price vs Rating Trend')
plt.xlabel('Final Price')
plt.ylabel('Rating')
plt.show()
```
