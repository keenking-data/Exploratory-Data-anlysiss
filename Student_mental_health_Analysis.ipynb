```python
print("The aim of this Explolatory Analysis is to find out if there exist an impact of a course  and marital status on a student mental health")
```

```python
#importing the neccesary libraries for this analysis
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#reading the student menthal health  dataset into the datafram data
data = pd.read_csv("/content/Student Mental health.csv")
data
```

```python
data.info()
```

```python
# to list all the columns present in the dataset
data.columns
```

```python
# plot to visualize the Depression Status Across Different Courses
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='What is your course?', hue='Do you have Depression?', palette='viridis')
plt.title('Depression Status Across Different Courses')
plt.xlabel('Course')
plt.ylabel('Number of Students')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Depression Status Across Different Courses** :

Generally courses like BCS and Engineering have the highest rate of depression .

Some courses like Islamic Education and banking Studies tend to have no depression at all a testament that students enrolled on this course find it enjoyable.
Contrarianly the course like BCS and also Enginneering have ahigher rate of non depression, meaning that the courses an almost diverse types of students on the pertaking of the coyurses and overally this courses might have he highest enrollment.  


### Marital Status vs. Depression

```python
# plot to visualize the Marital Status vs Depression Analysis
depression_proportion = data.groupby('Marital status')['Do you have Depression?'].value_counts(normalize=True).unstack().get('Yes', pd.Series(0, index=data['Marital status'].unique())).reset_index()
depression_proportion.columns = ['Marital status', 'Proportion_Depression_Yes']

plt.figure(figsize=(8, 6))
sns.barplot(data=depression_proportion, x='Marital status', y='Proportion_Depression_Yes', palette='plasma')
plt.title('Proportion of Students with Depression by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Proportion of Students with Depression')
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```

**Marital Status vs Depression Analysis** :




As per the analysis the students currently under marital status tend to have a higher proportion to be under depression as compared to students with no marital status

### Marital Status vs. Anxiety

```python
# Calculate the proportion of 'Yes' for Anxiety by Marital status
anxiety_proportion = data.groupby('Marital status')['Do you have Anxiety?'].value_counts(normalize=True).unstack().get('Yes', pd.Series(0, index=data['Marital status'].unique())).reset_index()
anxiety_proportion.columns = ['Marital status', 'Proportion_Anxiety_Yes']

plt.figure(figsize=(8, 6))
sns.barplot(data=anxiety_proportion, x='Marital status', y='Proportion_Anxiety_Yes', palette='viridis')
plt.title('Proportion of Students with Anxiety by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Proportion of Students with Anxiety')
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```

**Proportion of Students with Anxiety by Marital Status** :


The conclusion of the analysis is that students with marital status tend to be quite anxious more at a 0.5 rate as compared to students with marital status at 0.3 rate





### Marital Status vs. Panic Attack

```python
# plot to visualize  the Visualization of Panic attack vs Marital status
panic_attack_proportion = data.groupby('Marital status')['Do you have Panic attack?'].value_counts(normalize=True).unstack().get('Yes', pd.Series(0, index=data['Marital status'].unique())).reset_index()
panic_attack_proportion.columns = ['Marital status', 'Proportion_Panic_Attack_Yes']

plt.figure(figsize=(8, 6))
sns.barplot(data=panic_attack_proportion, x='Marital status', y='Proportion_Panic_Attack_Yes', palette='magma')
plt.title('Visualization of Panic attack vs Marital status')
plt.xlabel('Marital Status')
plt.ylabel('Proportion of Students with Panic Attack')
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```

**Visualization of Panic attack vs Marital status** :

The proportion of students currently with marital status tend to be experience higher rates of panic attack than students with no marital status.
